```python
import math
def find_mean(numbers):
    sum_numbers = 0
    for num in numbers:
        sum_numbers += num
    return sum_numbers / len(numbers)
    
def findSD(numbers):
  mean = find_mean(numbers)
  sum_squared_diff = 0
  for i in numbers:
    sum_squared_diff += (i - mean) ** 2
    variance = sum_squared_diff / (len(numbers) - 1)
    SD = math.sqrt(variance)
  return SD

def filter_numbers(numbers):
    mean = find_mean(numbers)
    SD = findSD(numbers)
    threshold = 1.5 * SD
    new_array = []
    for num in numbers:
        if abs(num - mean) >= threshold:
            new_array.append(num)
    return new_array



def ArrayProb(numbers):
  mean=find_mean
  SD=findSD(numbers)
  new_array=filter_numbers(numbers)



  





# This is the driver code
numbers = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
new_array=ArrayProb(numbers)
print(new_array)
unittest.output_test(new_array, np.array([25, -5]))
```
